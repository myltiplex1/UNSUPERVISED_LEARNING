```python
df_dummies = pd.get_dummies(df.drop('name',axis=1))
```

```python
df_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale = MinMaxScaler()
```

```python
scaled = scale.fit_transform(df_dummies)
```

```python
scaled
```

```python
scaled.max()
```

```python
scaled_df = pd.DataFrame(scaled, columns = df_dummies.columns)
```

```python
scaled_df
```

```python
sns.heatmap(scaled_df);
```

```python
sns.heatmap(scaled_df.corr(), annot = True);
```

```python
sns.clustermap(scaled_df,row_cluster=False);
```

```python
sns.clustermap(scaled_df,col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(data=df,x='mpg',y='weight',hue=cluster_labels)
```

```python
model = AgglomerativeClustering(n_clusters=None,distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10))
# Warning! This plot will take awhile!!
dn = hierarchy.dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(linkage_matrix,truncate_mode='lastp',p=48)
```
