```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_df = scaler.fit_transform(df)
```

```python
scaled_df
```

```python
scaled_data = pd.DataFrame(scaled_df, columns= df.columns)
```

```python
scaled_data
```

```python
from sklearn.cluster import DBSCAN
```

```python
outliers = []
for eps in np.linspace(0.001,3,50):
    model = DBSCAN(eps = eps, min_samples=2)
    model.fit(scaled_df)
    
    outliers.append(100 * (np.sum(model.labels_ == -1) / len(model.labels_)))
```

```python
sns.lineplot(x = np.linspace(0.001,3,50), y = outliers)
plt.xlabel('Epsilon values')
plt.ylabel('Percentage of Outliers');
```

```python
sns.lineplot(x = np.linspace(0.001,3,50), y = outliers)
plt.ylim(0,4)
plt.hlines(y=3,xmin=0,xmax=3,colors='red',ls='--')
plt.xlabel('Epsilon values')
plt.ylabel('Percentage of Outliers');
```
